```python
from llama_index.core import VectorStoreIndex, Settings, load_index_from_storage
from llama_index.core.storage.docstore import SimpleDocumentStore
from llama_index.core import StorageContext
from llama_index.core.node_parser import SentenceWindowNodeParser
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core.vector_stores import SimpleVectorStore
import os

# Define the persistence directory to save index metadata
persist_dir = "./storage"
os.makedirs(persist_dir, exist_ok=True)

# Check if we need to create and store the index for the first time
if not os.path.exists(os.path.join(persist_dir, "docstore.json")):
    # Define the text splitter and node parser
    text_splitter = SentenceSplitter()

    # Set global settings
    Settings.text_splitter = text_splitter

    # Create the sentence window node parser
    node_parser_sentence = SentenceWindowNodeParser.from_defaults(
        window_size=3,
        window_metadata_key="window",
        original_text_metadata_key="original_text",
    )

    # Create SimpleDocumentStore and SimpleVectorStore
    docstore = SimpleDocumentStore()
    vector_store = SimpleVectorStore()

    # Create storage context with these stores
    storage_context = StorageContext.from_defaults(
        docstore=docstore,
        vector_store=vector_store
    )

    # Process your documents (assuming documents are defined)
    nodes_sentence = node_parser_sentence.get_nodes_from_documents(documents)

    # Create and store the sentence index
    sentence_index = VectorStoreIndex(
        nodes_sentence,
        storage_context=storage_context,
        embed_model=embed_model,  # Make sure embed_model is defined
        llm=llm,  # Make sure llm is defined
        show_progress=True,
        use_async=True
    )

    # Persist the index to disk
    sentence_index.storage_context.persist(persist_dir)

    print("Index created and stored successfully!")
else:
    # Load the existing index from storage
    storage_context = StorageContext.from_defaults(
        persist_dir=persist_dir
    )

    sentence_index = load_index_from_storage(
        storage_context=storage_context,
        embed_model=embed_model,  # Make sure embed_model is defined
        llm=llm  # Make sure llm is defined
    )

    print("Index loaded from existing storage!")

# Create retriever that you can use multiple times
retriever_sentence = sentence_index.as_retriever(similarity_top_k=3)

# Now you can use the retriever in different sessions or applications
```

```python
from llama_index.core.postprocessor import LongContextReorder
# from llama_index.core.prompts import ChatPromptTemplate

from llama_index.core.prompts import RichPromptTemplate

custom_rich_prompt_str = """
{% chat role="system" %}
You are a specialized AI assistant that only answers questions about AI for class 6,7,8,9 students of CBSE board.
If a question is about another topic, politely explain that you can only provide information about AI.
Please provide a factual answer in simple words, in depth and with real life examples.
{% endchat %}
{% chat role="user" %}
The following is some retrieved context:

<context>
{{ context_str }}
</context>

Using the context, answer the provided question:
{{ query_str }}
{% endchat %}
"""

text_qa_template = RichPromptTemplate(custom_rich_prompt_str)

reorder = LongContextReorder()




# Create the query engine using the combined retriever
final_query_engine = RetrieverQueryEngine.from_args(
    retriever=combined_retriever,
    response_mode="tree_summarize",
    text_qa_template=text_qa_template,
    node_postprocessors=[reorder],# Optional: Choose response mode
)

# response = final_query_engine.query("I am a class 6th student. Explain automation and AI")
response = final_query_engine.query("I am a class 6th student. Explain automation and AI")

from llama_index.core.response.notebook_utils import display_response
display_response(response)

```
